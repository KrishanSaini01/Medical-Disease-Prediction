```python
df = df.drop_duplicates()
```

```python
for col in df.columns:
    print("\nColumn:", col)
    print(df[col].unique())
```

```python

Gender_list = df['Gender'].unique().tolist()
Gender_dict = {Gender: i for i, Gender in enumerate(Gender_list)}
print(Gender_dict)
df['Gender'] = df['Gender'].map(Gender_dict)


df['Smoking'] = df['Smoking'].map({'No':0,'Yes':1}).astype(int)

df['Alcohol'] = df['Alcohol'].map({ 'No':0,'Yes':1})

df['FamilyHistory'] = df['FamilyHistory'].map({'No':0,'Yes':1})

df['PhysicalActivity'] = df['PhysicalActivity'].map({'Medium':2, 'High':3, 'Low':1})

df['Disease'] = df['Disease'].map({'Healthy':1, 'Pre-Diabetes':2, 'Hypertension':3, 'Heart Disease':4, 'Diabetes':5})
```

```python
df.to_csv("medical_disease_cleaned.csv",index=False)
```

```python
x = df.drop('Disease', axis=1)
y = df['Disease']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=42,stratify=y)
```

# SVM (SVR)

```python
from sklearn.svm import SVR
sv=SVR()
sv.fit(x_train,y_train)
```

```python
print('SVR test score:-',sv.score(x_test,y_test))
print('SVR train score:-',sv.score(x_train,y_train))
```

# SVM (SVC)

```python
from sklearn.svm import SVC
sv=SVC()
sv.fit(x_train,y_train)
```

```python
print('SVC test score:-',sv.score(x_test,y_test))
print('SVC train score:-',sv.score(x_train,y_train))
```

# LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
log =LogisticRegression()
log.fit(x_train,y_train)
```

```python
print("LogisticRegression training",log.score(x_train,y_train))
print("LogisticRegression testing",log.score(x_test,y_test))
```

# DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
dt =DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

```python
print('DecisionTreeClassifier test score:-',dt.score(x_test,y_test))
print('DecisionTreeClassifier train score:-',dt.score(x_train,y_train))
```

# with parameters

```python
from sklearn.tree import DecisionTreeClassifier
dt =DecisionTreeClassifier(criterion='entropy', min_samples_split=3, min_samples_leaf=2, random_state=42)
dt.fit(x_train, y_train)
```

```python
print('DecisionTreeClassifier test score:-',dt.score(x_test,y_test))
print('DecisionTreeClassifier train score:-',dt.score(x_train,y_train))
```

# RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier 
import joblib
rfr =RandomForestClassifier()
rfr.fit(x_train, y_train)
```

```python
print(' test score:-',rfr.score(x_test,y_test))
print('train score:-',rfr.score(x_train,y_train))
```

```python
joblib.dump(rfr, 'RandomForestClassifier.lb')
print("Model saved successfully!")
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier,KNeighborsRegressor
knn=KNeighborsClassifier()
knn.fit(x_train, y_train)
```

```python
print("KNeighborsRegressor Test Score:", knn.score(x_test, y_test))
print("KNeighborsRegressor Train Score:", knn.score(x_train, y_train))
```
